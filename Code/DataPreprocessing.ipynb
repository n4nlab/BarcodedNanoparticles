```python
import pandas as pd
```

```python
# import barcodes data as dataframes

nBarcodes = 26 # introduce here the number of barcodes (csv files to read)
barcodes = [] # create iterable list to save the dataframes
for i in range(1,nBarcodes+1):
    if i<10: # changing 1 to 01
        j=str(i) #cast int to string
        j=j.rjust(2, '0') # add left 0 to the number to make it match the format
    else:
        j=str(i)
    name = "../_DataResults/FinalResults_Compiled/final_results_complete_%s.csv" % j # add counter to the filename
    print(name)
    barcodes.append(pd.read_csv(name))  #read csv and import it as a dataframe  
```

```python
# Drop columns (comment later)

for i in range(0,nBarcodes):   
    barcodes[i] = barcodes[i].drop(columns=['ch3_int'])
    
barcodes[3]
```

```python
# homogenize samples and export dataFrames to csv

for i in range(1,nBarcodes+1):
    if len(barcodes[i-1])>2000: # select only 2000 particles maximum for each dataset
        barcodes[i-1] = barcodes[i-1].sample(frac=1.0).head(2000) #shuffle and select first 2000
        
    if i<10: # changing 1 to 01
        j=str(i) #cast int to string
        j=j.rjust(2, '0') # add left 0 to the number to make it match the format
    else:
        j=str(i)
        
    name = "../_DataResults/FinalResults_Compiled/final_balanced_%s.csv" % j # add counter to the filename

    balanced = barcodes[i-1]
    balanced.to_csv(name) # export
```

```python
# split and export dataFrames to csv | 80% training 20% testing

for i in range(1,nBarcodes+1):
    if len(barcodes[i-1])>2000: # select only 2000 particles maximum for each dataset
        barcodes[i-1] = barcodes[i-1].sample(frac=1.0).head(2000) #shuffle and select first 2000
        
    if i<10: # changing 1 to 01
        j=str(i) #cast int to string
        j=j.rjust(2, '0') # add left 0 to the number to make it match the format
    else:
        j=str(i)
        
    name = "FinalResults_Compiled/final_training_dropped_%s.csv" % j # add counter to the filename
    nameTest = "FinalResults_Compiled/final_testing_dropped_%s.csv" % j # add counter to the filename
    
    train = barcodes[i-1].sample(frac=0.8, random_state=200) # randomly split 80% of the data
    train.to_csv(name) # export
    
    test = barcodes[i-1].drop(train.index) # save data to test dataset except the ones used in the training dataset
    test.to_csv(nameTest) # export
```
